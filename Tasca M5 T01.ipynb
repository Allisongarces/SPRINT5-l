```python
datos = pd.read_csv("LligueseuropeesUEFA.csv", sep=";", encoding="ISO-8859-1", index_col="Rk")

df = pd.DataFrame(data=datos)
df.head(10)
```

```python
#Ejercicio 1

print('Media de la columna MP "Juegos Jugados:"',df['MP'].mean())

tset, pval = ttest_1samp(df['MP'], 20)

print('P-valor',pval)

if pval < 0.05:    #Valor Alpha is 0.05 or 5%
   print('Se rechaza la hipotesis nula')
else:
  print('Se acepta la hipotesis nula')
```

```python
#Ejercicio 2
df[['xGD','xGD/90']].describe()

# H0: - Hay diferencia entre las medias
# H1: . No hay diferencia entre las medias

ttest,pval = stats.ttest_rel(df['xGD'], df['xGD/90']) 

print('P-valor:',pval)

if pval<0.05: 
    print("Rechazar hipótesis nula") 
else: 
    print("Aceptar hipótesis nula")
```

```python
from scipy.stats import f_oneway

#No se puede rechazar H0 : todas las distribuciones de muestra son iguales.
#Rechazar H0 : una o más distribuciones de muestra no son iguales.

stat, p = f_oneway(df['GF'],df['GA'],df['GD'])
print('Estadísticas=%.3f, p=%.3f' % (stat,p))
print('P-valor:',p)

alpha = 0.05

if p > alpha:
 print('Rechazar hipótesis nula')
else:
 print('Aceptar hipótesis nula')
```
